# Intro ML

Machine Learning può essere considerato una sottocategoria dell'AI. Sono algoritmi fondamentali per permettere ai computer di imparare e comportarsi in maniera più intelligente. Lo scopo è **generalizzare** anziché recuperare e salvare dati come fanno fanno i database.

Creiamo programmi con parametri che possiamo gestire, raffinare, e che sono poi aggiustati automaticamente per migliore e adattarsi a dati che non hanno visto.

Generalizziamo, per ottenere una **risposta** - non una soluzione - al nostro quesito.

## Perché e Quando usare ML?

### Perché


- **Automatizzare** compiti ripetitivi o complessi
- **Predire** dati futuri basandosi su dati già disponibili

### Quando?


- **Quando c'è un pattern** nei dati
- **Quando non c'è una regola** che possiamo scrivere a mano
- **Quando non sappiamo** come scrivere la regola
- **Quando i dati sono troppi o troppo complessi** per analizzarli manualmente

## Procedura generale


- è un problema da risolvere con ML?
- che dati abbiamo?
- *quanti* dati abbiamo?
- cosa possiamo dedurre dai dati? (esplorare i dati)
- ci sono tutti i dati che mi servono? sono troppi (troppe variabili)?

## Esempi

### Iris

Iris è un genere di piante della famiglia delle Iridaceae. Il genere comprende circa 300 specie di piante erbacee perenni, diffuse in Europa, Africa e Asia.

Sono facili da coltivare, e hanno dei colori brillanti.

La loro classificazione è un problema stocio in statistica e ML.

- **Problema:** Classificare le specie di iris
- **Dati:** Lunghezza e larghezza del sepalo e del petalo
- **Soluzione:** Dobbiamo classificare in base a questi dati.

### Diabete

Il diabete è una malattia cronica caratterizzata da un'eccessiva presenza di glucosio nel sangue. Può essere causato da fattori genetici, ambientali o comportamentali. La diagnosi precoce è fondamente per prevenire complicazioni. Predire la prograssione di diabete basandoci su misure cliniche è un problema comune in ML.

- **Problema:** Predire la progressione di diabete nel paziente a distanza di un anno dalla misurazione iniziale
- **Dati:** Pressione sanguigna, età, BMI, ecc...
- **Soluzione:** Predire la progresione del livello di glucosio basandoci su queste misure cliniche e su dati storici di pazienti simili

## Tipi di Apprendimento

La differenza tra l'esempio di Iris e quello di Diabete è tra *Classificazione* e *Regressione*.



### Classificazione

- **Output:** Variabili categoriche
> I fiori di irisi sono divisi in specie e queste speicie sono le categorie. Un altro esempio è la classificazione di email (spam/non spam). Prevedere **se** un paziente ha il diabete è un problema di classificazione.



### Regressione



- **Output:** Variabili continue
> Prevedere **la progressione** di diabete è un esempio di regressione. Altri esempi sono il prezzo delle case, la temperatura in base a cambiamenti climatici, ecc.

### Supervisionato o Non Supervisionato?


#### Supervisionato

- **Dati:** coppie di input e output
- **Obiettivo:** Predire l'output per nuovi input

L'apprendimento supervisionato è un metodo di apprendimento automatico in cui l'algoritmo impara da dati etichettati. L'algoritmo impara da esempi passati che contengono sia input che output. L'obiettivo è imparare una regola generale che mappa input a output.

#### Non Supervisionato

- **Dati:** Solo input
- **Obiettivo:** Trovare pattern nei dati

L'apprendimento non supervisionato è un metodo di apprendimento automatico in cui un algoritmo impara da dati non etichettati. L'algoritmo impara da esempi passati che contengono solo input. L'obiettivo è trovare il pattern nei dati senza conoscere l'output.

### Analisi Descrittiva o Predittiva?



> - Analisi Descrittiva: è il processo in cui esploriamo e riassumiamo i dari per ottenere una migliore comprensione del problema.
> - Analisi Predittiva: è il processo di costruzione di modelli per fare previsioni basate sui dati.



## Valutazione del Modello

> Il processo in cui misuriamo le prestazioni di un modello di ML in base a metriche specifiche.



### Metriche di valutazione



- *Accuracy*: Percentuale di predizioni corrette
- *Precision*: Percentuale di predizioni positive
- *Recall*: Percentuale di positivi predetti correttamente
- *F1 Score*: Media armonica di precision e recall
- *ROC Curve*: Grafico di True Positive Rate vs False Positive Rate



### Come valutare il modello?



- Training Set: Dati usati per addestrre il modello
- Validation Set: Dati usati per regolare i parametri del modello
- Test Set: Dati usare per valutare le prestazioni del modello



### Problemi comuni



> errori che possono verificarzsi durante la valutazione del modello e che possono influenzare le prestazioni del modello

#### Overfitting e Underfitting

- Overfitting: Il modello è troppo complesso e si adatta troppo ai dati di addestramento - perde la capacità di generalizzare.
- Underfitting: Il modello è troppo semplice - non riesce a catturare il pattern nei dati.

#### Bias e Variance

- Bias: Errore dovuto a un'errata assunzione sull'algoritmo
- Variance: Errore dovuto alla sensibilità del modello ai dati di addestramento

## Feature Engineering


> Il processo di creazione di nuove variabili a partire da quelle esistenti, oppure la selezione o trasformazione di quelle esistenti. Lo scopo è migliorare le prestazioni del modello, ridurre il ruomore nei dati, e facilitare l'interpretazione dei risultati.

- Feature Selection: Selezionare le variabili più rilevanti per il modello
- Feature Extraction: Creare nuove variabili partendo da quelle esistenti
- Feature Scaling: Standardizzare le variabili per facilitare l'addestramento del modello

## Algortimi di ML


> Sono metodi matematici e statistici che permettono ai computer di imparare dai dati e fare previsioni senza essere esplicitmente programmati.

Qualche esempio di algoritmi di ML:
- Linear Regression
- Logistic Regression
- Decision Trees
- Random Forest
- Support Vector Machines (AI)

### Trovare un equilibrio



#### Generalizzazione vs Memorizzazione

Un problema dell'apprendimento automatico è trovare un equilibrio tra generalizzazione e memorizzazione. 

Se memorizziamo troppo i dati di addestramento, il modello non sarà più in grado di generalizzare bene su unovi dati. Se generalizziamo tropppo, il modello potrebbe non essere più in grado di catturare il pattern nei dati.

Possiamo dividere il dataset in dati di addestramento e dati di test per valutare il modello.

#### Adestramento Supervisionato o Non supervisionato?

Formalmente:
Supervisionato: abbiamo un set di dati etichettati `X` in colonne e righe, dove colonne sono le variabili e le righe sono le osservazioni. Abbiamo anche un set di etichette `y` che corrispondono alle righe `X`. (Caso in cui sappiamo già come classificare i dati)
Non Supervisionato: **NON** abbiamo le etichette `y`, quindi dobbiamo trovare dei pattern nei dati `X` senza conoscere l'output. (Caso in cui non sappiamo bene quali sono i pattern o quando non abbiamo etichette) - ESEMPIO D'USO: la riduzione della dimensionalità (ridurre il numero di variabili in `X` senza perdere troppe informazioni, togliendo quelle superflue)

#### Regressione o Classificazione?

Possiamo vedere che la regressione è più adatta per prevedere la progressione del diabete, mentre la classificazione è più adatta per prevedere la specie di iris.



```python
from sklearn.model_selection import train_test_split

X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB

models = {
    'LogisticRegression': LogisticRegression(),
    'DecisionTree': DecisionTreeClassifier(),
    'KNeighbors': KNeighborsClassifier(),
    'LinearDiscriminant': LinearDiscriminantAnalysis(),
    'GaussianNB': GaussianNB()
}

from sklearn.metrics import accuracy_score

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_predicted = model.predict(X_test)
    results[name] = accuracy_score(y_test, y_predicted)

results
```

```python
# risultati
sns.boxplot(data=pd.DataFrame(results, index=[0]))
```

```python
classifier = KNeighborsClassifier()
classifier.fit(X_train, y_train)
y_predicted = classifier.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(accuracy_score(y_test, y_predicted))
print(confusion_matrix(y_test, y_predicted))
print(classification_report(y_test, y_predicted))
```
